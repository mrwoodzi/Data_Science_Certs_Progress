```python
import pandas as pd 
data = {"Index": [1980, 1981, 1982, 1983, 1984], 
        'India': [8880, 8670, 8147, 7338, 5704], 
        'China': [5123, 6682, 3308, 1863, 1527]
       }
df = pd.DataFrame(data)
df.set_index('Index', inplace=True)

# Plot the DataFrame without the index column
df.plot(kind='line', legend=True)

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Comparison')

# Display the plot
plt.show()
```
