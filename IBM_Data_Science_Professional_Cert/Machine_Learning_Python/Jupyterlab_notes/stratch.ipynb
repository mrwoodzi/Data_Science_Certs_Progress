```python
import math

# Function to calculate the resultant force
def calculate_resultant_force(forces):
    # Initialize variables for resultant components
    R_x = 0
    R_y = 0

    # Iterate through each force
    for force in forces:
        magnitude, direction = force
        # Convert direction to radians
        angle = math.radians(direction)

        # Calculate horizontal and vertical components
        force_x = magnitude * math.cos(angle)
        force_y = magnitude * math.sin(angle)

        # Sum up components
        R_x += force_x
        R_y += force_y

    # Calculate magnitude and direction of the resultant force
    R = math.sqrt(R_x**2 + R_y**2)
    theta = math.degrees(math.atan2(R_y, R_x))

    return R, theta

# Forces: (magnitude, direction)
forces = [(10, 0), (20, 315), (4.5, 90), (3, 0)]

# Calculate resultant force
resultant_force, direction = calculate_resultant_force(forces)

# Print the result
print(f"Resultant Force: {resultant_force:.2f} Newtons at {direction:.2f} degrees")
```

```python
import math
import matplotlib.pyplot as plt

# Function to calculate the resultant force
def calculate_resultant_force(forces):
    # Initialize variables for resultant components
    R_x = 0
    R_y = 0

    # Lists to store individual force components for plotting
    force_components = {'A': [], 'B': [], 'C': [], 'D': []}

    # Iterate through each force
    for i, force in enumerate(forces):
        magnitude, direction = force
        # Convert direction to radians
        angle = math.radians(direction)

        # Calculate horizontal and vertical components
        force_x = magnitude * math.cos(angle)
        force_y = magnitude * math.sin(angle)

        # Sum up components
        R_x += force_x
        R_y += force_y

        # Store individual force components for plotting
        force_components[chr(65 + i)] = (force_x, force_y)

    # Calculate magnitude and direction of the resultant force
    R = math.sqrt(R_x**2 + R_y**2)
    theta = math.degrees(math.atan2(R_y, R_x))

    return R, theta, force_components

# Function to plot vector diagram
def plot_vector_diagram(forces, resultant_force, force_components):
    fig, ax = plt.subplots()

    # Plot individual forces
    for force, components in force_components.items():
        ax.quiver(0, 0, components[0], components[1], angles='xy', scale_units='xy', scale=1, label=f'Force {force}')

    # Plot resultant force
    ax.quiver(0, 0, resultant_force[0], resultant_force[1], angles='xy', scale_units='xy', scale=1, color='r', label='Resultant Force')

    # Set plot properties
    ax.set_xlim(-max(resultant_force[0], max([abs(comp[0]) for comp in force_components.values()])) - 1, max(resultant_force[0], max([abs(comp[0]) for comp in force_components.values()])) + 1)
    ax.set_ylim(-max(resultant_force[1], max([abs(comp[1]) for comp in force_components.values()])) - 1, max(resultant_force[1], max([abs(comp[1]) for comp in force_components.values()])) + 1)
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)
    ax.set_aspect('equal', adjustable='box')
    ax.legend()
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

    # Show the plot
    plt.show()

# Forces: (magnitude, direction)
forces = [(10, 0), (20, 315), (4.5, 90), (3, 205)]

# Calculate resultant force
resultant_force, direction, force_components = calculate_resultant_force(forces)

# Print the result
print(f"Resultant Force: {resultant_force:.2f} Newtons at {direction:.2f} degrees")

# Plot vector diagram
plot_vector_diagram(forces, resultant_force, force_components)
```

```python
import math
import matplotlib.pyplot as plt

# Function to calculate the resultant force
def calculate_resultant_force(forces):
    # Initialize variables for resultant components
    R_x = 0
    R_y = 0

    # Lists to store individual force components for plotting
    force_components = {'A': [], 'B': [], 'C': [], 'D': []}

    # Iterate through each force
    for i, force in enumerate(forces):
        magnitude, direction = force
        # Convert direction to radians
        angle = math.radians(direction)

        # Calculate horizontal and vertical components
        force_x = magnitude * math.cos(angle)
        force_y = magnitude * math.sin(angle)

        # Sum up components
        R_x += force_x
        R_y += force_y

        # Store individual force components for plotting
        force_components[chr(65 + i)] = (force_x, force_y)

    # Calculate magnitude and direction of the resultant force
    R = math.sqrt(R_x**2 + R_y**2)
    theta = math.degrees(math.atan2(R_y, R_x))

    return R, theta, force_components

# Function to plot vector diagram with colored forces
def plot_vector_diagram(forces, resultant_force, force_components):
    fig, ax = plt.subplots()

    # Define colors for each force
    colors = {'A': 'blue', 'B': 'green', 'C': 'orange', 'D': 'red'}

    # Plot individual forces with colors
    for force, components in force_components.items():
        ax.quiver(0, 0, components[0], components[1], angles='xy', scale_units='xy', scale=1, color=colors[force], label=f'Force {force}')

    # Plot resultant force
    ax.quiver(0, 0, resultant_force[0], resultant_force[1], angles='xy', scale_units='xy', scale=1, color='r', label='Resultant Force')

    # Set plot properties
    ax.set_xlim(-max(resultant_force[0], max([abs(comp[0]) for comp in force_components.values()])) - 1, max(resultant_force[0], max([abs(comp[0]) for comp in force_components.values()])) + 1)
    ax.set_ylim(-max(resultant_force[1], max([abs(comp[1]) for comp in force_components.values()])) - 1, max(resultant_force[1], max([abs(comp[1]) for comp in force_components.values()])) + 1)
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_aspect('equal', adjustable='box')
    ax.legend()
    plt.grid(color='gray', linestyle='--', linewidth=0.5)

    # Show the plot
    plt.show()

# Forces: (magnitude, direction)
forces = [(10, 0), (20, 315), (4.5, 90), (3, 205)]

# Calculate resultant force
resultant_force, direction, force_components = calculate_resultant_force(forces)

# Print the result
print(f"Resultant Force: {resultant_force:.2f} Newtons at {direction:.2f} degrees")

# Plot vector diagram with colored forces
plot_vector_diagram(forces, resultant_force, force_components)
```

```python
def calculate_derivative(f, x, h=1e-5):
    """
    Calculate the numerical derivative of a function at a given point.
    
    :param f: The function.
    :param x: The point at which to calculate the derivative.
    :param h: The step size for numerical differentiation.
    :return: The numerical derivative.
    """
    return (f(x + h) - f(x - h)) / (2 * h)

# Example usage:
def example_function(x):
    return x**2 + 3*x + 5

# Calculate the first derivative at x = 2
first_derivative_at_2 = calculate_derivative(example_function, 2)

# Calculate the second derivative at x = 2
second_derivative_at_2 = calculate_derivative(lambda x: calculate_derivative(example_function, x), 2)

print(f"The first derivative at x=2 is: {first_derivative_at_2}")
print(f"The second derivative at x=2 is: {second_derivative_at_2}")

```
