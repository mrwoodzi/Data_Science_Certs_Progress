```python
import numpy as np
import matplotlib.pyplot as plt

# Define a real-life use case: Modeling the cost of production for a small manufacturing business.
# Imagine a business where the cost of production (in dollars) depends on the number of units produced.

# Define the coefficients of the polynomial.
# In this scenario, the polynomial represents the cost of production as a function of the number of units produced.
# The coefficients are determined based on historical data and analysis.
coefficients = [0.02, -0.5, 25]

# Generate x values for the plot.
# Here, 'x' represents the number of units produced, ranging from 0 to 100 units.
x = np.linspace(0, 100, 100)  # Create an array of 100 points from 0 to 100 units.

import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as optimize

# Define the coefficients of the polynomial.
coefficients = [0.02, -0.5, 25]

# Define the cost function to minimize.
def cost_function(units_produced, coefficients):
    return np.polyval(coefficients, units_produced)

# Use optimization to find the sweet spot (minimum cost).
result = optimize.minimize_scalar(cost_function, args=(coefficients,))
sweet_spot = result.x
min_cost = result.fun

# Generate x values for the plot.
x = np.linspace(0, 100, 100)  # Create an array of 100 points from 0 to 100 units.

# Calculate y values (the cost of production) for each x value using numpy.polyval.
y = np.polyval(coefficients, x)

# Create the plot.
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Cost of Production', color='blue')
plt.xlabel('Number of Units Produced')
plt.ylabel('Cost (in dollars)')
plt.title('Cost of Production vs. Number of Units Produced')
plt.grid(True)
plt.legend()

# Plot the sweet spot on the graph.
plt.scatter(sweet_spot, min_cost, color='red', label=f'Sweet Spot: {sweet_spot} units')

# Add labels and legend.
plt.annotate('Minimum Cost', xy=(sweet_spot, min_cost), xytext=(sweet_spot - 10, min_cost + 100),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5'))
plt.legend()

# Print the results.
print(f"Sweet Spot (Number of Units Produced): {sweet_spot}")
print(f"Minimum Cost: {min_cost} dollars")

plt.show()
```
