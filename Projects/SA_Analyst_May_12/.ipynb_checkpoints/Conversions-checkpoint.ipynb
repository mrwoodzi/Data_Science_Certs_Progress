```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the path to the CSV file
#Laptop
csv_file_path = r"C:\Users\mrwoo\OneDrive\Desktop\SA_Interview\all_311_transformed.csv"

#Server
#csv_file_path = "/home/mrwoo/Documents/all_331_transformed.csv"

# Reset display options to default
pd.reset_option('display.max_columns')
pd.reset_option('display.max_rows')
pd.reset_option('display.max_colwidth')

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Display the DataFrame
df
```

```python
# Convert 'Council District' to 'District' and change its type to string
df.rename(columns={'Council District': 'District'}, inplace=True)
df['District'] = df['District'].astype(str)

# Convert date columns to date objects
date_columns = ['OPENEDDATETIME', 'SLA_Date', 'CLOSEDDATETIME', 'Report Starting Date', 'Report Ending Date']
df[date_columns] = df[date_columns].apply(pd.to_datetime)
```

```python
df.drop(columns='OBJECTDESC', inplace=True)
```

```python
column_names = {'REASONNAME' : 'Reason', 'TYPENAME' : 'Type', 'CASEID' : 'Case'}
df.rename(columns=column_names, inplace=True)
```

```python
df.columns
```

```python
columns_change = { 'OPENEDDATETIME' : 'Opened', 'CLOSEDDATETIME' : 'Closed', 'SLA_Date' : 'SLA'}
df.rename(columns=columns_change, inplace=True)
```

```python
#Converting District column to numeric
df['District'] = pd.to_numeric(df['District'], errors='coerce')
```

```python
# Define a mapping dictionary for numerical values to words
number_to_word = {
    0: 'Zero',
    1: 'One',
    2: 'Two',
    3: 'Three',
    4: 'Four',
    5: 'Five',
    6: 'Six',
    7: 'Seven',
    8: 'Eight',
    9: 'Nine',
    10: 'Ten'
}

# Use the map() function to replace numerical values with words
df['District'] = df['District'].map(number_to_word)
```
